```python
import matplotlib.pyplot as plt
from algs import grahamscan, jarvismarch
from random import randint, uniform
import time
```

```python
A = [[uniform(-5, 5), uniform(-5, 5)] for i in range(10)]
graham = grahamscan(A)
print(A)
print(graham)

fig, ax = plt.subplots()
plt.xlabel('Width')
plt.ylabel('Height')

for i in A:
    ax.plot(i[0], i[1], 'ro')
for i in range(len(graham) - 1):
    ax.plot((A[graham[i]][0], A[graham[i + 1]][0]), (A[graham[i]][1], A[graham[i + 1]][1]), marker='o', linestyle='-')

ax.plot((A[graham[-1]][0], A[graham[0]][0]), (A[graham[-1]][1], A[graham[0]][1]))

plt.show()
```
